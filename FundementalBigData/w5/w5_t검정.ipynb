```python
import pandas as pd
from scipy import stats
import numpy as np
df=pd.read_csv("Ashopping.csv", sep=',', encoding='CP949')

#
no_claim=df[df.==0]
df2=np.array(no_claim.)

claim=df[df.==1]
df3=np.array(claim.)
stats.bartlett(df2, df3)
```

```python
#1.  t   
print(stats.ttest_ind(df2, df3, equal_var=False))
print(' (0)  : ', no_claim..mean())
print(' (1)  : ', claim..mean())
```
